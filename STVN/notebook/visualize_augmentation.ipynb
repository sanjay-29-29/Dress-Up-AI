```python
import sys; sys.path.append('..')

from torch.utils.data import DataLoader
from dataset import VITONHDDataset

import matplotlib.pyplot as plt
```

```python
dataset = VITONHDDataset('../DATA/VITON-HD', 1024, 768, is_paired=True, is_test=False)
dataloader = DataLoader(dataset, 1)
```

```python
n = 5
fig, axs = plt.subplots(n, 5, figsize=(10, 10))

for i in range(n):
    d = dataset[i]

    for j, k in enumerate(['image', 'cloth', 'agn', 'agn_mask', 'image_densepose']):
        ax = axs[i][j]
        ax.set_xticks([])
        ax.set_yticks([])
        if k == 'agn_mask':
            im = 1 - d[k]
            ax.imshow(im, cmap='gray')
        else:
            im = (d[k] + 1) / 2
            ax.imshow(im)
```
